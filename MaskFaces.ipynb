```python
import cv2

import matplotlib.pyplot as plt
```

```python
#use path to haarcascade file in your system when performing on jupyter notebook
cascade_file_path = r'C:\Users\Aaditya\Desktop\ABCOM\Testing of Idenity Masking\haarcascade_frontalface_default.xml'
```

```python
face_cascade = cv2.CascadeClassifier(cascade_file_path)
```

```python
# while on jupiter notebook the path of the imag file can be given as input to imread function
image = cv2.imread(r"img1.jpg")
```

```python
# Converting image from BGR to RGB format for matplotlib display function
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

```python
def blur(image):
  face = face_cascade.detectMultiScale(image,scaleFactor = 1.2)
  # Bluring each face detected and saved as list in face 
  for (x,y,h,w) in face:
    img = image[y:y+h, x:x+w]
    img = cv2.GaussianBlur(img,(99,99), 20)
    image[y:y+h,x:x+w] = img
    return image
```

```python
result = blur(image)
```

```python
plt.imshow(result)
```

```python
#VideoCapture is a function used in cv2 to load videos
sample_vid = cv2.VideoCapture(r"C:\Users\Aaditya\Desktop\ABCOM\Testing of Idenity Masking\Sample Video 2020-07-24 at 12.17.30 PM.mp4")
```

```python
# use the read function on your video object to capture it frame by frame
while True: 
  # Capture frame-by-frame 
  _, frame = sample_vid.read() 
  if frame is None: 
    break 

  else: 
    # converting frame to gray 
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) 
    blur (frame) 
    cv2.imshow("Video",frame) 

  #press 'q' key to abort 
  if cv2.waitKey(1) & 0xFF == ord('q'): 
    break 

# When everything is done, release the capture 

sample_vid.release() 
cv2.destroyAllWindows()
```
